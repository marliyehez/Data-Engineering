```python
from kafka import KafkaConsumer
from time import sleep
import json
from google.cloud import storage
from google.oauth2 import service_account
```

```python
IP_ADDRESS = ""   #   change the IP here
topic = "topic1"

consumer = KafkaConsumer(
    topic,
    bootstrap_servers=[f'{IP_ADDRESS}:9092'],
    value_deserializer=lambda x: json.loads(x.decode('utf-8')))
```

```python
# Set the bucket name
bucket_name = 'tweet-data-bucket'

# Initialize the Cloud Storage client
credentials = service_account.Credentials.from_service_account_file('gcs-creator.json')
storage_client = storage.Client(credentials=credentials)

# Iterate each message
for id, c in enumerate(consumer):
    print(c.value)

    # Convert dictionary to JSON string
    json_data = json.dumps(c.value)

    # Set the filename
    filename = f'tweet_{id}.json'

    # Upload JSON data to Cloud Storage
    bucket = storage_client.bucket(bucket_name)
    blob = bucket.blob(filename)
    blob.upload_from_string(json_data)
```
